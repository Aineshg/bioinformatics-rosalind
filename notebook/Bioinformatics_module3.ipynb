````python
# Ask user to input two positive integers
a, b = map(int, input("Enter two positive integers (a b) with a < b: ").split())

# Initialize sum
odd_sum = 0

# Loop through numbers from a to b (inclusive)
for number in range(a, b + 1):
    if number % 2 != 0:  # Check if number is odd
        odd_sum += number

# Print the sum of odd numbers
print(odd_sum)
````

````python
input_file = "rosalind_ini5.txt"

# Read all lines from the input file
with open(input_file, "r") as infile:
    lines = infile.readlines()

# Automatically save even-numbered lines to "even_lines.txt"
with open("even_lines.txt", "w") as outfile:
    for i in range(1, len(lines) + 1):  # 1-based numbering
        if i % 2 == 0:
            outfile.write(lines[i - 1])

print("Even-numbered lines have been saved to 'even_lines.txt'.")
````

####**TASK P18.2** <font color= orange>**Complete Rosalind Problem #5  Dictionaries**</font>

````python
# Read the input string
s = input("Enter a string: ").strip()

# Split the string into words
words = s.split()

# Create a dictionary to count occurrences
word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# Print the word counts (order does not matter)
for word, count in word_count.items():
    print(word, count)
````
